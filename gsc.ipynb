```python
from pymongo import MongoClient
import requests
import asyncio
import numpy as np
import pandas as pd
client = MongoClient("mongodb://localhost:27017/")
if(client):
   print("connected")
db = client.Title
collection1 = db['gsc']
df = pd.DataFrame(list(collection1.find()),columns=['URLs','clicks','ctr','impressions',
                                                    'position','views','Pattern'])
print(df)
```
